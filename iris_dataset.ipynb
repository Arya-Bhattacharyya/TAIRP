```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
Iris=pd.read_csv('Iris.csv')
Iris.head()
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=Iris)
plt.title('Scatter Plot of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

plt.show()
```

```python
sns.pairplot(Iris, hue='Species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)

plt.show()
```

```python
X = Iris.drop('Species', axis=1)
y = Iris['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state = 0)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=10)
knn_classifier.fit(X_train, y_train)
```

```python
y_pred = knn_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

Now we predicting the species of the Iris based on the user's input data.

```python
knn_classifier.fit(X[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']], y)

def user_input():
    SepalLengthCm = float(input("Enter sepal length (cm): "))
    SepalWidthCm = float(input("Enter sepal width (cm): "))
    PetalLengthCm	 = float(input("Enter petal length (cm): "))
    PetalWidthCm = float(input("Enter petal width (cm): "))
    return [[SepalLengthCm, SepalWidthCm, PetalLengthCm	, PetalWidthCm]]

user_data = user_input()

predicted_species = knn_classifier.predict(user_data)
print(f"Predicted Species: {predicted_species[0]}")
```
